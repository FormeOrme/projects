```python
import numpy as np
import matplotlib.pyplot as plt

# Image dimensions
width, height = 1000, 200
# Position values from 0 to 1
x = np.linspace(0, 1, width)

# Linear gradients
linear_gradient1 = x
linear_gradient2 = 1 - x
linear_result = linear_gradient1 * linear_gradient2

# squared gradient
squared_gradient1 = x ** 5
squared_gradient2 = (1 - x) ** 5
squared_result = squared_gradient1 * squared_gradient2

# Square root gradients
sqrt_gradient1 = np.sqrt(x)
sqrt_gradient2 = np.sqrt(1 - x)
sqrt_result = sqrt_gradient1 * sqrt_gradient2

# Cosine gradients
cosine_gradient1 = 0.5 * (1 - np.cos(x * np.pi))
cosine_gradient2 = 0.5 * (1 + np.cos(x * np.pi))
cosine_result = cosine_gradient1 * cosine_gradient2
```

```python


# Create images
linear_image = np.tile(linear_result, (height, 1))
sqrt_image = np.tile(sqrt_result, (height, 1))
cosine_image = np.tile(cosine_result, (height, 1))

# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Linear
axes[0].imshow(linear_image, cmap="gray", vmin=0, vmax=1, aspect="auto")
axes[0].set_title("Linear Multiplication")
axes[0].axis("off")

# Square root
axes[1].imshow(sqrt_image, cmap="gray", vmin=0, vmax=1, aspect="auto")
axes[1].set_title("Square Root Multiplication")
axes[1].axis("off")

# Cosine
axes[2].imshow(cosine_image, cmap="gray", vmin=0, vmax=1, aspect="auto")
axes[2].set_title("Cosine Multiplication")
axes[2].axis("off")

plt.show()
```

```python
# Plot
plt.figure(figsize=(12, 8))

# Linear
plt.plot(x, linear_gradient1, label='Linear Gradient 1', color='gray', linestyle='--', alpha=0.5)
plt.plot(x, linear_gradient2, label='Linear Gradient 2', color='gray', linestyle=':', alpha=0.5)
plt.plot(x, linear_result, label='Linear (Gradient 1  Gradient 2)', color='gray', linestyle='-')

# Squared
plt.plot(x, squared_gradient1, label='Squared Gradient 1', color='red', linestyle='--', alpha=0.5)
plt.plot(x, squared_gradient2, label='Squared Gradient 2', color='red', linestyle=':', alpha=0.5)
plt.plot(x, squared_result, label='Squared (Gradient 1  Gradient 2)', color='red', linestyle='-')

# Square root
plt.plot(x, sqrt_gradient1, label='Square Root Gradient 1', color='blue', linestyle='--', alpha=0.5)
plt.plot(x, sqrt_gradient2, label='Square Root Gradient 2', color='blue', linestyle=':', alpha=0.5)
plt.plot(x, sqrt_result, label='Square Root (Gradient 1  Gradient 2)', color='blue', linestyle='-')

# Cosine
plt.plot(x, cosine_gradient1, label='Cosine Gradient 1', color='green', linestyle='--', alpha=0.5)
plt.plot(x, cosine_gradient2, label='Cosine Gradient 2', color='green', linestyle=':', alpha=0.5)
plt.plot(x, cosine_result, label='Cosine (Gradient 1  Gradient 2)', color='green', linestyle='-')

plt.title('Gradient Multiplication: Linear, Squared, Square Root, and Cosine Compensation')
plt.xlabel('Position (x)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.1)
plt.xlim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
